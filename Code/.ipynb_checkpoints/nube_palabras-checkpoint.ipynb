```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
df= pd.read_csv(r'C://Users//juan.romero//Downloads//sessions_with_survey.csv')
```

```python
text = " ".join(df['text'])
```

```python
# Lista de stop words adicionales
additional_stopwords = {'de','del', 'es', 'la', 'y','o', 'en', 'un', 'una', 'el', 'con','que','lo','los','las','para','por','al','te', 'a', 'su','si','no','este','tu', 'incluyendo','Porvenir','porvenir','PORVENIR'}

# Actualizar la lista de stop words del wordcloud con las stop words adicionales
stopwords = STOPWORDS.union(additional_stopwords)
```

```python
# Crear una instancia de WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords, collocations=False).generate(text.lower())

# Mostrar la imagen generada
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # No mostrar ejes
plt.show()
```
